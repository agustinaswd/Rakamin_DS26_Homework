4. Lakukan filter dengan ketentuan sebagai berikut: (6pts)

    - Tidak ada nilai null pada column `occupation_type`
    - Memiliki value pada column `occupation_type` dengan pola `staff`.
5. Sudah dilakukan sorting berdasarkan `income_type` dari `A-Z` dan `amount_income_total` dari `besar ke kecil`. (6pts)

**Expected Result:**

![Expected Result](assets/homework_exercise_3_result.png)

```python
## dataframe baru dari dataset application_processed.csv

#1
#flag_email = 1
data1 = data_app_process[data_app_process['flag_email']==1]
data1

#cek duplikat email
data1.duplicated().any()
data1[data1.duplicated(keep=False) == True]

#dataframe baru dengan flag_email = 1 yang sudah dihapus data yang duplikat
data2 = data1.drop_duplicates()
data2.duplicated().any()
```

```python
#2
#column baru
data2.loc[(data2['flag_own_car']==1) & (data2['flag_own_car']==1), 'has_car_and_property']=1
data2.loc[(data2['flag_own_car']==0) & (data2['flag_own_car']==1), 'has_car_and_property']=0
data2.loc[(data2['flag_own_car']==1) & (data2['flag_own_car']==0), 'has_car_and_property']=0
data2.loc[(data2['flag_own_car']==0) & (data2['flag_own_car']==0), 'has_car_and_property']=0
data2
```

```python
#3
#column yang diperlukan
data3 = data2[['id', 'gender', 'has_car_and_property', 'income_type', 'education_type', 'family_status', 'housing_type', 'occupation_type', 'amount_income_total']]
data3
```

```python
#4
#filter
data4 = data3[(data3['has_car_and_property']==1) & (data3['occupation_type'].notnull()) & (data3['occupation_type'].str.contains('staff'))] 
data4
```

```python
#5
#sorting berdasarkan asc income_type & amount_income_total desc
data5 = data4.sort_values(['income_type', 'amount_income_total'], ascending=[True,False])
data5
```

```python
#6
#export
data5.to_csv('homework_3_result.csv', index=False)
pd.read_csv('homework_3_result.csv')
```

### 5. Memanfaatkan Pivot Tabel (15pts)

Buatlah dataframe baru menggunakan fungsi pivot table untuk melihat `Embarked` apa saja yang `average_female_ticket_price` lebih besar daripada `average_male_ticket_price`. Informasi `average_female_ticket_price` dan `average_male_ticket_price` didapatkan dari rata-rata pada column `Fare`.

**Hint:**
1. menggunakan pivot table
2. dapat dilihat pada topik `DataFrame Aggregation` bagian `Pivot Table | Reshape Columns and Rows`

**Expected Result:**

![Expected Result](assets/homework_exercise_5_result.png)

```python
## DataFrame baru dengan pivot table
#group by 'Fare'
titanic2 = titanic1.groupby(['Embarked', 'Sex']).agg({'Fare' : ['mean']})
titanic2 = titanic2.reset_index()
titanic2.columns = ['Embarked', 'Sex', 'Fare']
titanic2

#pivot table
titanic3 = pd.pivot_table(titanic2,
                          index = ['Embarked'],
                          columns = ['Sex'],
                          values = ['Fare']
                         )
titanic3 = titanic3.reset_index()
titanic3.columns = ['Embarked', 'avg_female_ticket_price', 'avg_male_ticket_price']
titanic3[titanic3['avg_female_ticket_price'] > titanic3['avg_male_ticket_price']]
```

### 7. Merge DataFrame (25pts)

Buatlah 2 dataframe mengikuti petunjuk berikut:

df_1 :  Filter dataset 01 telecom_revenue.csv dengan multiple kondisi (5pts)
- MonthlyRevenue > 10 
- Occupation terdiri dari Professional, Student, and Crafts

df_2 :  Filter dataset 02 telecom_usage.csv dengan multiple kondisi (5pt)
- nilai UnansweredCalls > .BlockedCalls

Jawablah pertanyaan berikut ini:
1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

**Expected Result:**

![Expected Result](assets/homework_exercise_7_result.png)

```python
## Your code below here
df1 = pd.read_csv('01 telecom_revenue.csv')

#filter
df_1 = df1[(df1['MonthlyRevenue'] > 10) & (df1['Occupation'].isin(['Professional', 'Student', 'Crafts']))]
df_1
```

```python
#df_2

df2 = pd.read_csv('02 telecom_usage.csv')
df2

#filter
df_2 = df2[(df2['UnansweredCalls']>df2['BlockedCalls'])]
df_2
```

```python
#merge df

df_3 = df_1.merge(df_2,
                  on = 'CustomerID'
                 )
df_3
```

```python
#df aggregation
df_4 = df_3.groupby(['Occupation']).agg({'CustomerID': 'nunique', 'DroppedCalls' : 'mean'})
df_4
```

```python
jadi, 
- total CustomerID dari masing-masing Occupation adalah Professional = 128 dengan rata-rata DroppedCalls = 5.54, Student = 5 dengan rata-rata DroppedCalls = 8.22
- rata-rata Occupation terbesar ada pada Occupation Student
```
